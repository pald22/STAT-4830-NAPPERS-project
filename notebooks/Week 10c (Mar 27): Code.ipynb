# Polymarket mispricing portfolio (baseline)

This notebook runs the modular pipeline documented in `docs/polymarket_mispricing_pipeline.md`.

**Setup:** add the repo `src` directory to `PYTHONPATH` (e.g. in the first code cell) so `polymarket` imports resolve.

```python
from pathlib import Path
import sys

ROOT = Path.cwd()
REPO = ROOT.parent if ROOT.name == "notebooks" else ROOT
sys.path.insert(0, str(REPO / "src"))

from polymarket.backtest import MispricingConfig, run_walk_forward
from polymarket.diagnostics import plot_run
from polymarket.load import enrich_snapshot, find_markets_csv, load_polymarket_snapshot

raw = load_polymarket_snapshot(find_markets_csv(REPO))
df = enrich_snapshot(raw)
df.shape
```

```python
from pathlib import Path

REPO = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
OUT = REPO / "outputs" / "polymarket_mispricing"
cfg = MispricingConfig(n_folds=5, ridge_alpha=5.0, gamma=2.0, kappa=0.05)
result = run_walk_forward(df, cfg=cfg, out_dir=OUT)
result["summary"]
```

```python
plot_run(result, out_dir=OUT)
print("Saved:", OUT / "diagnostics.png")
```
